## Field of view

We want to relate the field of view of the microscope in epifluorescence mode to the field of view in 2-photon mode.

```python
from scipy.misc import imread

epi_img = imread('imgs/epi-fov.jpg', flatten=True)
tp_img = tiff.imread('imgs/Sys5_20140721_1338_FOV/1/FOV000001.tif')

spec = gs.GridSpec(8,8)
fig = plt.figure(figsize=(6,8))

epi = fig.add_subplot(spec[:4, :])
epi.imshow(epi_img)
epi.get_xticklabels()
epi.get_yticklabels()
epi.set_title('epifluorescence')

tp1 = fig.add_subplot(spec[4:, :4])
tp1.imshow(tp_img[0])
tp1.set_xticklabels([])
tp1.set_yticklabels([])
tp1.set_title('2p, channel 1')

tp2 = fig.add_subplot(spec[4:, 4:])
tp2.imshow(tp_img[1])
tp2.set_xticklabels([])
tp2.set_yticklabels([])
tp2.set_title('2p, channel 2')
```

We will use 2p channel 1 for all further analyses. 

###### Binarize the images

```python
import scipy.ndimage as spndi

tp1_img = tp_img[0]
sig = 4
epi_gf = spndi.gaussian_filter(epi_img, sig)
tp1_gf = spndi.gaussian_filter(tp1_img, sig)

# binarize
epi_thresh = epi_gf.mean() + 3.5 * epi_gf.std()
tp1_thresh = tp1_gf.mean() + 3.5 * tp1_gf.std()
epi_b = np.where(epi_gf < epi_thresh, 0, 1)
tp1_b = np.where(tp1_gf < tp1_thresh, 0, 1)

epi_lb, elabels = spndi.label(epi_b)
tp1_lb, tlabels = spndi.label(tp1_b)

epi_cm = np.array(spndi.center_of_mass(epi_b, epi_lb, range(1, elabels+1)))
# we notice that we have a label we don't want to include going forward, we will
# manually exclude it below
tp1_cm = np.array(spndi.center_of_mass(tp1_b, tp1_lb, range(1, tlabels))) 

fig, (epi, tp1) = plt.subplots(2, 1)
epi.imshow(epi_lb)
epi.plot(epi_cm[:, 1], epi_cm[:, 0], 'r+')
epi.set_xticklabels([])
epi.set_yticklabels([])
epi.set_xlim(0, epi_lb.shape[1])
epi.set_ylim(epi_lb.shape[0], 0)
epi.set_title('epifluorescence')

tp1.imshow(tp1_lb)
tp1.plot(tp1_cm[:, 1], tp1_cm[:, 0], 'r+')
tp1.set_xticklabels([])
tp1.set_yticklabels([])
tp1.set_xlim(0, tp1_lb.shape[1])
tp1.set_ylim(tp1_lb.shape[0], 0)
tp1.set_title('2p, channel 1')
```

Now that we have center points for each pollen grain, we will attempt to match the pollen grains represented in the 2p image with pollen grains in the epifluorescence image. 

1. Choose three pollen grains from the 2p image.
2. Calculate the triangle formed by those pollen grains. 
3. Iteratively calculate all possible triangles in the epi image, comparing to the 2p triangle.

```python
import itertools

def _unit_vector(a):
    return a / np.linalg.norm(a)

def angle_between(a, b):
    a = _unit_vector(a)
    b = _unit_vector(b)
    return np.arccos(np.clip(np.dot(a, b), -1, 1))

def dist(a, b):
    dp = b - a
    return np.sqrt(np.sum(dp ** 2))

class Triangle(object):
    
    def __init__(self, a, b, c):
        self.a = a
        self.b = b
        self.c = c
        angle_a = angle_between(b - a, c - a)
        angle_b = angle_between(c - b, a - b)
        angle_c = angle_between(a - c, b - c)
        self.ang = np.sort([angle_a, angle_b, angle_c])
        self.dists = np.array([dist(a, b), dist(a, c), dist(b, c)])
        self.dists = np.sort(self.dists / np.max(self.dists))
        
    def difference(self, other):
        return np.sum(np.abs(self.ang - other.ang) + np.abs(self.dists - other.dists))
    
    def angle_offset(self, other):
        pass
        
    def scale_diff(self, other):
        pass
    
# 2p triangle
tp_triangle = Triangle(*tp1_cm[:3])

# all others
diffs = []
for x in itertools.combinations(xrange(elabels), 3):
    # get points
    pts = (epi_cm[x[0]], epi_cm[x[1]], epi_cm[x[2]])
    tri = Triangle(*pts)
    diff = tp_triangle.difference(tri)
    diffs.append((diff, pts))

min_diff = min(diffs, key=lambda x: x[0])
fig, epi = plt.subplots(1, 1)
epi.imshow(epi_lb)
pts = np.array(min_diff[1])
epi.plot(pts[:, 1], pts[:, 0], 'r+')
epi.set_xticklabels([])
epi.set_yticklabels([])
epi.set_xlim(0, epi_lb.shape[1])
epi.set_ylim(epi_lb.shape[0], 0)
epi.set_title('minimum diff triangle points, diff : '+str(min_diff[0]))
```

From this, we can see that the y-axis in the 2p FOV is reflected, and that the FOV itself stretches downward and to the right away from the center point of the epi FOV.

## Microns per pixel

We found the microscope to have approximately 450 microns horizontally and approximately 430 microns vertically, which, given the typical 512x512 resolution, works out to $450 / 512 = .879$ microns per pixel horizontally and $430 / 512 = .840$ microns per pixel vertically.
